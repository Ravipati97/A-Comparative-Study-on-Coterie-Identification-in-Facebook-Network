```python
import os
import os.path
import gzip
import re
import numpy as np
```

```python
def read_gz_file(path):  
    if os.path.exists(path):  
        with gzip.open(path, 'r') as pf:  
            for line in pf:  
                yield line  
    else:  
        print('the path [{}] is not exist!'.format(path))
```

```python
fpath = "./facebook/"
egos = ['0','107','348','414','686','698','1684','1912','3437','3980']
nodeNum = 4039
```

```python
###  feature name and its id
featureMap = {}
for ego in egos:
    featureFile = fpath+ego+".featnames"
    f = open(featureFile,'r')
    for line in f:
        #print line
        m = re.split('(\d+)',line)
        category = m[2][1:-20]
        feat = int(m[-2])
        if category not in featureMap:
            #print category
            featureMap[category] = [feat]
        elif feat not in featureMap[category]:
            featureMap[category].append(feat)
    f.close()
```

```python
### number of feature
lst = []
for feats in featureMap.values():
    lst+=feats
sizeFeature = max(lst)+1
#print len(set(lst))
#print max(lst)
```

```python
###   same feature id, different feature name
featureMap_reverse={}
for f in featureMap:
    vec = featureMap[f]
    for v in vec:
        if v not in featureMap_reverse:
            featureMap_reverse[v]=[f]
        else: featureMap_reverse[v].append(f)
featureMap_reverse
```

```python
###  get feature of each node
featureOfNode = {}
for i in range(nodeNum):
    featureOfNode[i]=[0]*sizeFeature

for ego in egos:
    featureFile = fpath+ego+".featnames"
    fMap = []
    f = open(featureFile,'r')
    for line in f:
        m = re.split('(\d+)',line)
        feat = int(m[-2])
        fMap.append(feat)
    f.close()
    
    featureOfNodeFile = fpath+ego+".feat"
    f = open(featureOfNodeFile,'r')
    for line in f:
        m = line.split()
        node = int(m[0])
        for i in range(1,len(m)):
            if (m[i]=='1'):
                featureOfNode[node][fMap[i-1]] = 1
    f.close()
    
    featureOfEgoFile = fpath+ego+".egofeat"
    f = open(featureOfEgoFile,'r')
    for line in f:
        m = line.split()
        node = int(m[0])
        for i in range(1,len(m)):
            if (m[i]=='1'):
                featureOfNode[node][fMap[i-1]] = 1
    f.close()
```

```python
### write to file
lines = []
for i in range(sizeFeature):
    s = ""
    s += str(i)
    for feat in featureOfNode[i]:
        s = s+' '+str(feat)
    s+='\n'
    lines.append(s)
featureOfNodeFile = fpath+'feature.txt'
f = open(featureOfNodeFile,'w')
f.writelines(lines)
f.close()
```
